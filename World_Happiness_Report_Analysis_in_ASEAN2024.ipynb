# World Happiness Report
นั้นมักจะอิงจากแหล่งข้อมูลหลายประเภท ซึ่งมีทั้งข้อมูลทางสถิติที่วัดจากแหล่งข้อมูลทางเศรษฐกิจและสังคม รวมถึงข้อมูลที่ได้จากการสำรวจความคิดเห็นของประชาชนในแต่ละประเทศ นี่คือตัวอย่างวิธีการเก็บข้อมูลสำหรับปัจจัยสำคัญต่าง ๆ:

###Ladder Score (คะแนนความสุข):

ที่มา: เป็นผลมาจากการสำรวจความคิดเห็นของผู้คนในแต่ละประเทศ โดยใช้แบบสอบถามที่เรียกว่า "Cantril Ladder" โดยให้ผู้ตอบให้คะแนนความสุขส่วนตัวในช่วงจาก 0 ถึง 10 ซึ่ง 0 คือสถานการณ์ที่แย่ที่สุดที่พวกเขาสามารถจินตนาการได้ และ 10 คือสถานการณ์ที่ดีที่สุด
วิธีการเก็บข้อมูล: ข้อมูลนี้ได้จากการสำรวจประชาชนตัวอย่างหลายพันคนในแต่ละประเทศ โดยใช้แบบสอบถามมาตรฐาน เช่น Gallup World Poll
###Log GDP per capita (ผลิตภัณฑ์มวลรวมต่อหัว):

ที่มา: ข้อมูลนี้มาจากการวัดเศรษฐกิจของประเทศ โดยคำนวณ GDP (Gross Domestic Product) หารด้วยจำนวนประชากร
วิธีการเก็บข้อมูล: ข้อมูลนี้มาจากแหล่งข้อมูลทางเศรษฐกิจระดับสากล เช่น ธนาคารโลก (World Bank), กองทุนการเงินระหว่างประเทศ (IMF) หรือองค์การเพื่อความร่วมมือทางเศรษฐกิจและการพัฒนา (OECD)
###Social Support (การสนับสนุนทางสังคม):

ที่มา: ข้อมูลนี้มาจากการสำรวจประชาชนเกี่ยวกับการรับการสนับสนุนทางสังคม โดยถามว่าในสถานการณ์ที่ยากลำบากหรือวิกฤต พวกเขาสามารถพึ่งพาคนในชุมชนหรือครอบครัวได้หรือไม่
วิธีการเก็บข้อมูล: ใช้แบบสอบถาม เช่นเดียวกับ Gallup World Poll
###Healthy Life Expectancy (อายุขัยที่คาดหวังอย่างมีสุขภาพดี):

ที่มา: ข้อมูลนี้มาจากองค์กรระดับสากล เช่น องค์การอนามัยโลก (WHO) ซึ่งประเมินสุขภาพและอายุขัยเฉลี่ยของประชาชนในแต่ละประเทศ
วิธีการเก็บข้อมูล: ใช้สถิติทางสุขภาพ เช่น อัตราการตาย อัตราการเจ็บป่วย และข้อมูลอื่น ๆ ที่เกี่ยวข้อง
###Freedom to Make Life Choices (อิสรภาพในการตัดสินใจชีวิต):

ที่มา: มาจากการสำรวจถามผู้คนว่าพวกเขารู้สึกว่ามีอิสรภาพในการตัดสินใจเกี่ยวกับชีวิตของตนเองมากน้อยเพียงใด
วิธีการเก็บข้อมูล: ใช้แบบสอบถามเกี่ยวกับอิสรภาพทางเศรษฐกิจ สังคม และการเมือง
###Generosity (ความเอื้อเฟื้อเผื่อแผ่):

ที่มา: มาจากการถามผู้คนว่าพวกเขาบริจาคให้การกุศลหรือให้ความช่วยเหลือกับผู้อื่นมากน้อยเพียงใด
วิธีการเก็บข้อมูล: ใช้แบบสอบถามเกี่ยวกับการทำบุญ การบริจาค หรือการช่วยเหลือผู้อื่น
###Perceptions of Corruption (การรับรู้ถึงการคอร์รัปชัน):

ที่มา: ถามประชาชนว่าพวกเขารับรู้ว่ามีการคอร์รัปชันในรัฐบาลหรือภาคธุรกิจของประเทศมากน้อยเพียงใด
วิธีการเก็บข้อมูล: ใช้แบบสอบถามเกี่ยวกับความเชื่อมั่นในรัฐบาลและการบริหารงาน
###Dystopia + Residual (ค่าส่วนต่างที่อธิบายด้วยประเทศที่มีสถานการณ์แย่ที่สุด):

ที่มา: Dystopia เป็นการอ้างอิงถึงประเทศที่มีคะแนนความสุขต่ำที่สุดในโลก โดยตัวเลขนี้เป็นการคาดการณ์ว่า ถ้าทุกปัจจัยในประเทศนั้นอยู่ในสถานะที่แย่ที่สุด จะได้ค่าความสุขประมาณเท่าไหร่
วิธีการเก็บข้อมูล: คำนวณโดยการใช้โมเดลทางคณิตศาสตร์เพื่อหาความแตกต่างจากค่าจริงของแต่ละประเทศ

# การวิเคราะห์ความสัมพันธ์ (Correlation Analysis)
ความสัมพันธ์ระหว่าง GDP per capita กับ Ladder score ในกลุ่มอาเซียน:

```python
correlation_social_healthy_asean = asean_df['Explained by: Social support'].corr(asean_df['Explained by: Healthy life expectancy'])
print(f"Correlation between Social support and Healthy life expectancy in ASEAN countries: {correlation_social_healthy_asean}")
```

# กลุ่มประเทศที่มีค่าเฉลี่ยสูงสุดหรือต่ำสุดในอาเซียน


หาประเทศที่มีค่าเฉลี่ยของ Ladder score สูงที่สุดและต่ำที่สุดในกลุ่มอาเซียน:


```python
max_ladder_score_asean = asean_df.loc[asean_df['Ladder score'].idxmax()]
min_ladder_score_asean = asean_df.loc[asean_df['Ladder score'].idxmin()]

print("Country with highest Ladder score in ASEAN:")
print(max_ladder_score_asean[['Country name', 'Ladder score']])

print("\nCountry with lowest Ladder score in ASEAN:")
print(min_ladder_score_asean[['Country name', 'Ladder score']])
```

หาประเทศที่มีค่าเฉลี่ยของ Healthy life expectancy สูงที่สุดในกลุ่มอาเซียน:


```python
max_healthy_life_asean = asean_df.loc[asean_df['Explained by: Healthy life expectancy'].idxmax()]
print("Country with highest Healthy life expectancy in ASEAN:")
print(max_healthy_life_asean[['Country name', 'Explained by: Healthy life expectancy']])
```
